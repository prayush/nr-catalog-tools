```python
import pycbc.types as pt
import pycbc.waveform as wf
import matplotlib.pyplot as plt
```

```python
hp, hc = wf.get_td_waveform(
    approximant="SEOBNRv4",
    mass1=40,
    mass2=30,
    spin1z=0.5,
    f_lower=20,
    delta_t=1.0 / 4096,
)
```

```python
plt.plot(hp.sample_times, hp)
```

```python
my_real_data = hp.data
my_imag_data = -1 * hc.data
```

```python
my_real_data
```

```python
my_real_ts = pt.TimeSeries(my_real_data, delta_t=1.0 / 4096)
my_imag_ts = pt.TimeSeries(-1 * my_imag_data, delta_t=1.0 / 4096)
```

```python
plt.plot(my_real_ts.sample_times, my_real_ts)
```

```python
my_freq = wf.frequency_from_polarizations(my_real_ts, my_imag_ts)
```

```python
%matplotlib widget
plt.plot(my_freq.sample_times, my_freq)
```
